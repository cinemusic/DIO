**1 - Converter uma imagem colorida em níveis de cinza**

```python
!pip install Pillow
```

```python
# importação de bibliotecas necessárias
import PIL
from PIL import Image
import numpy as np
```

```python
# define constantes usadas no programa
PCT_R = 0.29 # percentual do nível de vermelho no cálculo da intensidade de um pixel
PCT_G = 0.59 # percentual do nível de verde no cálculo da intensidade de um pixel
PCT_B = 0.11 # percentual do nível de azul no cálculo da intensidade de um pixel
LIMIAR = 160 # nível de limiar para binarizar uma imagem em níveis de cinza
BRANCO = 255 # nível de cinza para a cor BRANCA na binarização, o nível zero é preto 
```

```python
# leitura da imagem colorida (original)
imagemRGB = Image.open("./img/lenna.jpg")
imagemRGB
```

```python
# separação da imagem, nos seus canais R, G e B
imagemR, imagemG, imagemB = imagemRGB.split()
```

```python
# cria array novo para imagem de níveis de cinza
arrayCinza = np.zeros([imagemRGB.width, imagemRGB.height], dtype = np.uint8)

# cria arrays para cada uma das imagens R, G e B
arrayR = np.array(imagemR)
arrayG = np.array(imagemG)
arrayB = np.array(imagemB)

# varre pixels do array da imagem em níveis de cinza e calcula a intensidade para cada um
# a fórmula para a intensidade de um pixel é: nivel = 0.29*R + 0.59*G + 0.11*B
for col in range(imagemRGB.width):
  for lin in range(imagemRGB.height):
    arrayCinza[lin, col] = int(arrayR[lin,col]*PCT_R) + int(arrayG[lin,col]*PCT_G) + int(arrayB[lin,col]*PCT_B)

# Cria imagem para o array de níveis de cinza e salva
imagemCinza = Image.fromarray(arrayCinza)
imagemCinza.save("./img/lenna_cinza.jpg")
imagemCinza
```

**2 - Conversão de uma imagem em níveis de cinza para binária**

```python
# Lê imagem em nível de cinza, salva na atividade anterior
imagemCinza2 = Image.open("./img/lenna_cinza.jpg")
imagemCinza2
```

```python
arrayBinario = np.zeros([imagemCinza2.width, imagemCinza2.height], dtype = np.uint8)
arrayCinza2 = np.array(imagemCinza2)

# varre o array da imagem em níveis de cinza e calcula a intensidade de cada um
# Se o nível for maior que o limiar (160), atribui o nível 255 (BRANCO) ao pixel
# Como o array foi inicializado com zeros, não precisa atribuir o nível 0 (PRETO)
for col in range(imagemCinza2.width):
  for lin in range(imagemCinza2.height):
    if arrayCinza2[lin, col] > LIMIAR:
      arrayBinario[lin, col] = BRANCO

# Cria imagem para o array de valores binários e salva
imagemBinario = Image.fromarray(arrayBinario)
imagemBinario.save("./img/lenna_binario_160.jpg")
imagemBinario
```
